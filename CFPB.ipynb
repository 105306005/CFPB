### 2-3-1 Set up different approach for each column to deal with missing values

Fill the rows where categorical values are missing with "missing", \

```python
# Fill the column with missing
top_5_bank_data["Sub-product"].fillna("missing", inplace=True)
top_5_bank_data["Sub-issue"].fillna("missing", inplace=True)
top_5_bank_data["Company public response"].fillna("missing", inplace=True)
top_5_bank_data["State"].fillna("missing", inplace=True)
top_5_bank_data["ZIP code"].fillna("missing", inplace=True)
top_5_bank_data["Tags"].fillna("missing", inplace=True)
top_5_bank_data["Consumer consent provided?"].fillna("missing", inplace=True)
top_5_bank_data["Company response to consumer"].fillna("missing", inplace=True)

# Drop the rows with missing in the "Consumer disputed?" column
top_5_bank_data.dropna(subset=["Consumer disputed?"], inplace=True)
```

### 2-4 Make sure it's all numerical & define our `X` and `y` 

- We want to turn all columns into numbers

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

top_5_bank_data_clean = pd.read_csv("raw_data/top_5_bank_data_clean.csv")
top_5_bank_data_clean.head()
```

```python
X = top_5_bank_data_clean[[ 'Product', 'Sub-product', 'Issue', 'Sub-issue',
        'Company public response', 'Company',
       'State',  'Tags', 'Consumer consent provided?',
       'Submitted via',  'Company response to consumer',
       'Timely response?']]
```

```python
# Next, we convert all the X observations into numbers
X = pd.get_dummies(X)
X
```

```python
y = top_5_bank_data_clean['Consumer disputed?']
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(top_5_bank_data_clean['Consumer disputed?'])
print(le.classes_)
print(le.inverse_transform(list(set(y))))
labels = dict(zip(list(set(y)), le.inverse_transform(list(set(y)))))
print("\nLabel mapping is:", labels)
```

### 2-5 Split into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 2-6 Deal with imbalanced dataset - undersampling

```python
# from imblearn.over_sampling import SMOTE 
# smote = SMOTE()
# X_train, y_train = smote.fit_resample(X_train, y_train)

# undersampling gives us better results in the modeling section later on.

from imblearn.under_sampling import RandomUnderSampler
undersampler = RandomUnderSampler()
X_train, y_train = undersampler.fit_resample(X_train, y_train)
```

## 3 Choosing a model

## 3-1 XGBoost -----------

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, plot_confusion_matrix
from sklearn.metrics import mean_absolute_error, mean_squared_error


from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm

import sklearn.preprocessing as preproc
```

```python
# Fit the model
from xgboost import XGBClassifier

model_xgb = XGBClassifier(use_label_encoder=False, objective= 'binary:logistic')
model_xgb.fit(X_train, y_train)
```

```python
# Perform predictions, and store the results in a variable called 'pred'
pred = model_xgb.predict(X_test)
```

```python
# Check the classification report and the confusion matrix
print(classification_report(y_true = y_test, y_pred = pred))
plot_confusion_matrix(model_xgb, X = X_test, y_true = y_test, cmap='Greys');
```

This means we can play around with the threshold that decides what is marked as 'yes'. Feel free to reduce this threshold on the next cell, and see if it helps you get better recall for category 1 (that may reduce accuracy, but that is okay).

```python
# Set threshold for identifying class 1
threshold = 0.4

# Create predictions.  Note that predictions give us probabilities, not classes!
pred_prob = model_xgb.predict_proba(X_test)

# Convert probabilities to 1s and 0s based on threshold
pred = (pred_prob>threshold).astype(int)

pred = np.argmax(pred, axis = 1) # argmax converts the 2 probabilities (one for each class) into a single prob defined by threshold

# confusion matrix
cm = confusion_matrix(y_test, pred)
print ("Confusion Matrix : \n", cm)
 
# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, pred))
print('Test recall = ', metrics.recall_score(y_test, pred, average="macro"))
```

```python
print(classification_report(y_true = y_test, y_pred = pred))
```

### AUC and ROC calculation - logistic regression 

```python
# AUC calculation
metrics.roc_auc_score(y_test, pred_prob)
```

```python
# Source for code below: https://stackoverflow.com/questions/25009284/how-to-plot-roc-curve-in-python
fpr, tpr, thresholds = metrics.roc_curve(y_test, pred_prob)
roc_auc = metrics.auc(fpr, tpr)
plt.figure(figsize = (9,8))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
